# Clean phantom fraction workflow
This notebook demonstrates the refactored GitHub package workflow. The complete original notebook is stored separately.


```python
from ml_phantom.config import DatasetConfig
from ml_phantom.data import load_dataset, save_dataset_npz, split_dataset
from ml_phantom.models.architectures import build_model3, compile_model
from ml_phantom.metrics import evaluate_model
```

## 1. Load and preprocess data


```python
config = DatasetConfig(base_path=r'C:/Users/Mahdieh/NK/ML/phantom', main_folders=[f'S{i}' for i in range(1, 10)])
images, labels, records = load_dataset(config)
print(images.shape, labels.shape)
save_dataset_npz('data/processed/phantom_dataset.npz', images, labels, records)
```

## 2. Split data and build Model 3


```python
X_train, X_test, y_train, y_test = split_dataset(images, labels, test_size=0.2, random_state=42)
model = build_model3(input_shape=X_train.shape[1:], num_phases=y_train.shape[-1], grid_cells=y_train.shape[1])
model = compile_model(model, use_decay=True)
model.summary()
```

## 3. Train


```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32)
model.save('outputs/models/model3.keras')
```

## 4. Evaluate


```python
metrics = evaluate_model(model, X_train, y_train, X_test, y_test, phase_names=config.phase_names)
metrics
```
